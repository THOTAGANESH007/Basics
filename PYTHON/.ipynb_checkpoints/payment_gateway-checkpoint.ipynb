# Functions for the payment simulation

```python
# Possible payment outcomes
PAYMENT_OUTCOMES = [
    ("succeeded", 0.65),
    ("failed", 0.20),
    ("timeout", 0.10),
    ("insufficient_funds", 0.05),
]

# Allowed Methods
allowed_methods = ["upi", "card", "cash"]
```

```python
# To simulate the payment status
def get_random_status():
    rand_prob = random.random()
    cumulative = 0
    for outcome, prob in PAYMENT_OUTCOMES:
        cumulative += prob
        if rand_prob <= cumulative:
            return outcome
```

```python
# Generate the payment_id
def payment_number_generation(amount, method):
    payment_id = uuid.uuid4().hex[:12]
    print(f"Payment Id: {payment_id}")
    print(f"Amount:{amount} || Method:{method}")
    return payment_id
```

```python
def process_payment(payment_id):
    print("Payment Processing...")
    time.sleep(random.uniform(1,3))
    status = get_random_status()
    return status
```

```python
def max_tries(tries, max_tries, payment_id):
    while tries <= max_tries:
        status = process_payment(payment_id)
        match status:
            case "succeeded":
                print("Payment Success!!!")
                return
            case "failed":
                print("Payment Failed!!!")
                return
            case "insufficient_funds":
                print("Insufficient Funds to proceed Further!!!")
                return
            case "timeout":
                print("Timed Out (Network Error)")
                tries += 1
                if tries > max_tries:
                    return
                do_you_continue = input("Enter any key to retry or N/n to stop: ")
                if do_you_continue.casefold() == "n":
                    return
            case _:
                print("...")
```

```python
try:
    def simulate_payment():
        amount = int(input("Enter Amount to withdraw: "))
        method = input("Enter Method Name (Upi/Card/Cash): ").casefold()
        if method not in allowed_methods:
            raise Exception("Please Enter the valid Payment Method")
            
        payment_id = payment_number_generation(amount, method)

        tries = 0
        MAX_TRIES = 3
        max_tries(tries, MAX_TRIES, payment_id)
except ValueError as e:
    print(e)
except Exception as e:
    print(e)

simulate_payment()
```
